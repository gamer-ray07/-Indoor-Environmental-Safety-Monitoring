```python
anomaly_labels = iso_model.predict(X)
```

```python
import pandas as pd
pd.Series(anomaly_labels).value_counts()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
normal=df[anomaly_labels== 1]
anomaly=df[anomaly_labels== -1]
plt.scatter(normal["temp"],normal["co"],alpha=0.3)
plt.scatter(anomaly["temp"],anomaly["co"],alpha=0.8)

plt.xlabel("Temprature")
plt.ylabel("CO Level")
plt.title("Isolation Forest Anomaly Detection")
plt.show()
```

```python
normal_data = df[anomaly_labels == 1]
anomaly_data = df[anomaly_labels == -1]

print("Normal CO mean:", normal_data["co"].mean())
print("Anomaly CO mean:", anomaly_data["co"].mean())

print("Normal CO max:", normal_data["co"].max())
print("Anomaly CO max:", anomaly_data["co"].max())
```

```python
print("Normal temp mean:", normal_data["temp"].mean())
print("Anomaly temp mean:", anomaly_data["temp"].mean())
```

```python
print("Normal humidity mean:", normal_data["humidity"].mean())
print("Anomaly humidity mean:", anomaly_data["humidity"].mean())
```
